```python
Here each element of [0, 0, 1, 1] represents:

The first value (0): This represents the left edge of the axes, where 0 corresponds to the leftmost edge of the figure, and 1 corresponds to the rightmost edge.

The second value (0): This represents the bottom edge of the axes, where 0 corresponds to the bottom edge of the figure, and 1 corresponds to the top edge.

The third value (1): This represents the width of the axes. In this case, it's set to 1, which means the axes will span the entire width of the figure.

The fourth value (1): This represents the height of the axes. Setting it to 1 means the axes will span the entire height of the figure.

So, ax = fig.add_axes([0, 0, 1, 1]) creates a set of axes that covers the entire figure, essentially creating a single subplot that takes up the entire figure space. You can then use ax to plot your data within this subplot.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data
x = np.linspace(0, 10, 100)       # 100 points between 0 and 10
y = np.sin(x)                     # sine values for line and scatter plot
categories = ['A', 'B', 'C']      # categories for bar chart
values = [5, 7, 3]                # values for bar chart
data = np.random.randn(1000)      # random data for histogram

# Create subplots
fig, axes = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = axes.flatten()

# Plot different types of charts
ax1.plot(x, y, color='blue', label='Line Plot')
ax1.legend()

ax2.scatter(x, y, color='red', label='Scatter Plot')
ax2.legend()

ax3.bar(categories, values, color='green')
ax3.set_title('Bar Chart')

ax4.hist(data, bins=30, color='purple')
ax4.set_title('Histogram')

# Adjust layout and show
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig,axes = plt.subplots(1,3, figsize = (12,4))
x=np.arange(1,11)
axes[0].plot(x, x**3, 'g',lw=2)
axes[0].grid(True)
axes[0].set_title('default grid')
axes[1].plot(x, np.exp(x), 'r')
axes[1].grid(color='b', ls='-.',lw=0.25)
axes[1].set_title('custom grid')
axes[2].plot(x,x)
axes[2].set_title('no grid')
fig.tight_layout()
plt.show()
```

```python
Matplotlib  Twin Axes
when plotting curves with different units together

the plot has dual y axes, one showing exp(x) and the other showing log(x):
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig=plt.figure()
a1=fig.add_axes([0,0,1,1])
x=np.arange(1,11)
a1.plot(x,np.exp(x))
a1.set_label('exp')
a2=a1.twinx()
a2.plot(x, np.log(x),'ro-')
a2.set_ylabel('log')
fig.legend(labels=('exp','log'), loc=4)
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['category A', 'category B', 'category C', 'category D']
sizes = [25, 30, 15, 20]
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen' ]
explode = (0.2, 0.2, 0.2, 0.2)

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, pctdistance=0.85, explode=explode)
center_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(center_circle)
plt.axis('equal')
plt.title('Donut Chart')
plt.show()
```

```python
import pandas as pd

# Your data
data = {
    'Name': ['rohan','mohan','sohan','gulshan','roy'],
    'roll_no': [6,8,9,3,5],
    'python_marks': [56,78,90,46,67],
    'powerbi_marks': [96,78,60,46,67],
    'excel_marks': [76,78,90,96,67]
}

df = pd.DataFrame(data)

# Calculate total marks and percentage
df['total_marks'] = df[['python_marks','powerbi_marks','excel_marks']].sum(axis=1)
df['percentage'] = (df['total_marks'] / 300) * 100

print(df[['Name','roll_no','total_marks','percentage']])
```

```python
#wap to check year is leap year or not
```

```python
year=int(input("enter year"))

if year%4==0:
    if year%100==0:
        if year%400==0:
            print(year,"is leap  year")
        else:
            print(year,"not leap year")
    else:
        print(year,"is leap year")
else:
    print(year,"not leap year")
```

```python
amount=int(input("enter total amount of purchase in rs:-"))
product=input("enter product that u have taken (gold/silver/other)").lower()
payment_mode=input("enter payment mode (upi/cash)").lower()

if amount>=50000:
    if product=="gold":
        discount=0.30
    elif product=="silver":
        discount=0.20
    else:
        discount=0.10
else:
    if product=="gold":
        discount=0.10
    else:
        discount=0.0

final_amount=amount-(amount*discount)

if payment_mode=="upi":
    final_amount=final_amount-final_amount*0.05
    print(final_amount)
else:
    print(final_amount)
    
    
```

```python
# find the greatest common divisor
```

```python
def find_gcd(a,b):
    return math.gcd(a,b)
find_gcd(84,16)
```

```python
def compound_interest(p,r,n,t):
    return p*math.pow(1+(r/n),n*t)
compound_interest(10000,0.05,12,3)
```

```python
# find lcm of any number
```

```python
def lcm(a,b):
    return abs(a*b)//math.gcd(a,b)
lcm(19,18)
```

```python
def eucledian_distance(x1,y1,x2,y2):
    return math.hypot(x2-x1,y2-y1)
eucledian_distance(1,4,4,5)
```

```python
def scalene_Area(a,b,c):
    s=(a+b+c)/2
    return math.sqrt(s*(s-a)*(s-b)*(s-c))
scalene_Area(3,4,5)
```

```python
def infinite_geometric_sum(a,r):
    if abs(r)<1:
        return a/(a-r)
    else:
        return None
infinite_geometric_sum(10,0.5)
```

```python
def permutation(n,r):
    return math.factorial(n)//math.factorial(n-r)
permutation(10,2)
```

```python
def combination(n,r):
    return math.comb(n,r)
permutation(10,2)
```
